# Linear Regression MVP

My goal is to predict % of stadium capacity (attendance [defined by ticket sales] / specific stadium capacity) filled. I am looking for a focus on predictive capabilities over interpretability. Knowing what drives attendance from a team performance point of view does not help us improve ticket sales, as it would likely not go over well to tell a team to 'play better' so we sell more tickets. So the focus is on alternative methods for selling more tickets, such as promotions, social media campaigns, etc.

My initial features were:
   * Weekend - Binary
   * max_temp - Float 
   * precip - Binary
   * night - Binary (defined by baseball reference)
   * opp_WL - Float [0-1]
   * avg_runs [over past 5 games] - Float
   * avg_runs_allowed [over past 5 games] - Float
   * streak - Int
   * WL_pct - Float [0-1]
   * games_back - Int (negative implies ahead of 2nd place team)
   


There was some multicollinearity between some of the team performance features (VIF seen above). I removed some of the features that were contributing to the high VIF, and was left with: Weekend, precip, night, avg_runs, avg_runs_allowed, WL_pct, games_back. I then created a linear regression model with these features. The predicted vs. residuals plot can be seen below.



The model has a significant F-stat and the coeffecient values can be seen below, along with their p-values. The relationship between the predicted y-hat and actuals doesn't appear completely linear, and I am exploring the potential for transforming y.
